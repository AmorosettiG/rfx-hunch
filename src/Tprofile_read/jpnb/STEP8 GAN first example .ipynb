```python
gan = models.GAN0.GAN0(latent_dim=2, feature_dim=30, dprate=0.3, scale=1)
```

```python
class AsyncTrain(ThreadWithLogAndControls):
    def __init__(self, model, data, **kwargs):
        self._model = model
        model.stop_training = False
        fn = lambda thread_print: model.train(data, **kwargs)
        super(AsyncTrain, self).__init__(target=fn, name='async train')
    
    def terminate(self):
        self._model.stop_training = True
```

```python
task = AsyncTrain(gan, qsh, epoch=100, batch=100, optimizer_factor=1e-3)
task.control_panel()
```
